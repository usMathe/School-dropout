**História 4**

NARRATIVA: Como administrador posso avaliar a situação dos alunos internacionais e nacionais separadamente.

CRITÉRIOS DE ACEITAÇÃO:
- Devem ser gerados um relatório para cada tipo de aluno(nacional ou internacional).
- deve ser apresentado um gráficos de pizza para cada tipo;
- o grafico deve deve ser apresentado em formato percentual a quantidade de abandono e sucesso para cada um dos tipos.

```python
# Criar um dicionário para armazenar as contagens de desistências para cada nacionalidade
dropout_counts = {}

# Percorrer cada linha no dataframe
for index, row in df2.iterrows():
  # Obter a nacionalidade e os valores de saída
  nationality = row["internacional"]
  output = row["Output"]

  # Se a nacionalidade ainda não estiver no dicionário, adicione-a com uma contagem de 0
  if nationality not in dropout_counts:
    dropout_counts[nationality] = 0

  # Se a saída for "Dropout", aumente a contagem para essa nacionalidade
  if output == "Dropout":
    dropout_counts[nationality] += 1

# Imprimir as contagens de desistências para cada nacionalidade
for nationality, count in dropout_counts.items():
  filtered_df = df2[df2["internacional"] == nationality]

  # Obter o número total de estudantes para a nacionalidade atual
  total_estudentes = filtered_df.shape[0]

  # Calcular a taxa de abandono
  dropout_porcentage = (count / total_estudentes) * 100

  labels = ["Dropout", "Graduate"]
  sizes = [dropout_porcentage, 100 - dropout_porcentage]
  plt.pie(sizes, labels=labels, autopct="%1.1f%%",colors=['aqua', 'blue', 'steelblue'])
  if nationality == 1:
    plt.title(f"Taxa de abandono e graduação para estudante internacionais")
  else:
    plt.title(f"Taxa de abandono e graduação para estudante nacionais")
  plt.show()
```
